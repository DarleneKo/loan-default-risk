```python
import pandas as pd

df = pd.read_csv("credit_risk_dataset.csv")
df.head()
```

```python
df1 = df.dropna()
```

```python
data_with_loanstatus_sliced = df1[(df1['cb_person_default_on_file']=="Y") | (df1['cb_person_default_on_file']=="N")]
di = {"Y":0, "N":1}   #converting target variable to boolean
Dataset_withBoolTarget1= data_with_loanstatus_sliced.replace({"cb_person_default_on_file": di})
Dataset_withBoolTarget1.head()
```

```python
data_with_loanstatus_sliced = Dataset_withBoolTarget1[(Dataset_withBoolTarget1['person_home_ownership']=="RENT") | (Dataset_withBoolTarget1['person_home_ownership']=="OWN") | (Dataset_withBoolTarget1['person_home_ownership']=="MORTGAGE")]
di = {"RENT":0, "OWN":1, "MORTGAGE":2}   #converting target variable to boolean
Dataset_withBoolTarget2= data_with_loanstatus_sliced.replace({"person_home_ownership": di})
Dataset_withBoolTarget2.head()
```

```python
data_with_loanstatus_sliced = Dataset_withBoolTarget2[(Dataset_withBoolTarget2['loan_intent']=="DEBTCONSOLIDATION") | (Dataset_withBoolTarget2['loan_intent']=="EDUCATION") | (Dataset_withBoolTarget2['loan_intent']=="HOMEIMPROVEMENT") | (Dataset_withBoolTarget2['loan_intent']=="MEDICAL") | (Dataset_withBoolTarget2['loan_intent']=="PERSONAL") | (Dataset_withBoolTarget2['loan_intent']=="VENTURE")]
di = {"DEBTCONSOLIDATION":0, "EDUCATION":1, "HOMEIMPROVEMENT":2, "MEDICAL":3, "PERSONAL":4, "VENTURE":5}   #converting target variable to boolean
Dataset_withBoolTarget3= data_with_loanstatus_sliced.replace({"loan_intent": di})
Dataset_withBoolTarget3.head()
```

```python
data_with_loanstatus_sliced = Dataset_withBoolTarget3[(Dataset_withBoolTarget3['loan_grade']=="A") | (Dataset_withBoolTarget3['loan_grade']=="B") | (Dataset_withBoolTarget3['loan_grade']=="C") | (Dataset_withBoolTarget3['loan_grade']=="D") | (Dataset_withBoolTarget3['loan_grade']=="E") | (Dataset_withBoolTarget3['loan_grade']=="F") | (Dataset_withBoolTarget3['loan_grade']=="G")]
di = {"A":0, "B":1, "C":2, "D":3, "E":4, "F":5, "G":6}   #converting target variable to boolean
Dataset_withBoolTarget4= data_with_loanstatus_sliced.replace({"loan_grade": di})
Dataset_withBoolTarget4.head()
```

```python
Dataset_withBoolTarget4.to_csv("clean_credit_risk.csv", index=False)
```
